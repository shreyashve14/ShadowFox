# LINE PLOT

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
x = [1,2,3,4]
y = [10,20,30,40]
plt.plot(x, y,marker='o',color='blue')
plt.title("Line Plot-Matplotlib")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.grid(True)
plt.show()
df_line = pd.DataFrame({"x":x,"y":y})
sns.lineplot(data=df_line,x="x",y="y",marker="o",color='red')
plt.title("Line Plot-Seaborn")
plt.show()
```

# BAR PLOT

```python
categories = ['A', 'B', 'C']
values = [5, 7, 3]
plt.bar(categories, values, color='green')
plt.title("Bar Plot-Matplotlib")
plt.show()
df_bar = pd.DataFrame({'Category': categories, 'Value': values})
sns.barplot(x='Category', y='Value', data=df_bar)
plt.title("Bar Plot - Seaborn")
plt.show()
```

#  HISTOGRAM

```python
data_hist = [1, 2, 2, 3, 3, 3, 4, 4, 5]
plt.hist(data_hist, bins=5, color='orange', edgecolor='black')
plt.title("Histogram - Matplotlib")
plt.show()

# Seaborn
sns.histplot(data_hist, bins=5, color='green')
plt.title("Histogram - Seaborn")
plt.show()
```

# PIE CHART

```python
labels = ['A', 'B', 'C']
sizes = [40, 35, 25]
plt.pie(sizes, labels=labels, autopct='%1.1f%%',
        colors=['pink', 'lightblue', 'lightgreen'])
plt.title("Pie Chart - Matplotlib")
plt.show()
```
